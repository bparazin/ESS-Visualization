This is a quick & dirty workbook I wrote to put all 41 yearly files generated by the GloFAS trimming workbook into 1 file that I can use as my master data copy

```python
import netCDF4 as nc
import env_methods as em
import os
import datetime as dt
import pickle
```

```python
file_path = r'C:\Users\bpara\Geo data\GloFAS compressed'
```

```python
final_data = dict()
with open(os.path.join(file_path + '\\' + '1979mississippi data'), 'rb') as f:
    dat1979 = pickle.load(f)

final_data["lon"] = dat1979["lon"]
final_data["lat"] = dat1979["lat"]
final_data["time"] = dat1979["time"]
final_data["dis24"] = dat1979["dis24"]

for x in os.listdir(file_path)[1:]:
    with open(os.path.join(file_path + '\\' + x), 'rb') as f:
        temp = pickle.load(f)
        final_data["time"] += temp["time"]
        final_data["dis24"] += temp["dis24"]
   
```

```python
file = open("Full period data", "wb")
pickle.dump(final_data, file)
file.close()
```
