```python
import netCDF4 as nc
import env_methods as em
import os
import datetime as dt
import pickle
```

```python
main_path = r'C:\Users\bpara\Geo data'

_1979_path = r'\river 1979'
```

```python
def compress_year_data(filepath, start_day, year, filename):
    data = nc.Dataset(os.path.join(filepath, os.listdir(filepath)[0]))
    result = dict()
    
    lat_max = em.find_closest_val(27.98036759814649, data["lat"])
    lat_min = em.find_closest_val(50.381277, data["lat"])

    lon_min = em.find_closest_val(-114.914572, data["lon"])
    lon_max = em.find_closest_val(-75.78187893993685, data["lon"])
    
    
    result["lon"] = data["lon"][lon_min:lon_max]
    result["lat"] = data["lat"][lat_min:lat_max]
    result["time"] = [start_day]
    result["dis24"] = [data["dis24"][0, lat_min:lat_max, lon_min:lon_max][:]]
    data.close()
    
    current_day = start_day
    TIME_INTERVAL = dt.timedelta(days = 1)
    for day in os.listdir(filepath)[1:]:
        current_day += TIME_INTERVAL
        data = nc.Dataset(os.path.join(filepath, day))
        result["time"].append(current_day)
        result["dis24"].append(data["dis24"][0, lat_min:lat_max, lon_min:lon_max][:])
        data.close()
        
    file = open(filename, "wb")
    pickle.dump(result, file)
    file.close()
    
    return result
```

```python
year = 1999

start_day = dt.datetime(year, 1, 1)

compress_year_data(main_path + _1979_path, start_day, year, str(year) + "mississippi data")
```
